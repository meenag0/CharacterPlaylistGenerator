```python
def movie_music_genres(movie_genre):
    if isinstance(movie_genre, str):
        genres = movie_genre.split(", ")
        # Drop 'family' and 'animation' from the list of genres
        genres = [genre for genre in genres if genre.lower() not in ['animation']]
        music_genres = set()
        for genre in genres:
            if genre in genre_mapping:
                music_genres.update(genre_mapping[genre])
        return ",".join(music_genres)
    else:
        return ""
```

```python
# Assuming you have both `character_df` and `genre_average_metrics_df` DataFrames

# Step 1: Extract unique values from "Associated Music Genres" column of character_df
unique_genres = character_df['Associated Music Genres'].unique()

# Step 2: Create an empty DataFrame to store the average metrics for each genre combination
average_genre_metrics_df = pd.DataFrame(index=unique_genres, columns=genre_avg_metrics_df.columns[1:])

# Step 3: Iterate over each unique genre combination and calculate the mean metrics
for genre_combination in unique_genres:
    # Extract individual genres from the combination
    genres = genre_combination.split(',')
    
    # Initialize a list to store metric values for each genre combination
    genre_combination_metrics = []
    
    # Check if the genre combination is "Unknown"
    if genre_combination == "Unknown":
        # Use spaecific genres to replace NaN values
        specific_genres = ["pop", "rock", "r&b", "ambient", "hip hop"]
        for genre in specific_genres:
            if genre in genre_avg_metrics_df['track_genre'].values:
                genre_metrics = genre_avg_metrics_df.loc[genre_avg_metrics_df['track_genre'] == genre].iloc[:, 1:]
                genre_combination_metrics.append(genre_metrics)
    else:
        # Iterate over each genre in the combination
        for genre in genres:
            # Check if the genre exists in genre_average_metrics_df
            if genre in genre_avg_metrics_df['track_genre'].values:
                # Append the metrics for this genre to the list
                genre_metrics = genre_avg_metrics_df.loc[genre_avg_metrics_df['track_genre'] == genre].iloc[:, 1:]
                genre_combination_metrics.append(genre_metrics)
    
    # If there are metrics for at least one genre in the combination
    if genre_combination_metrics:
        # Concatenate the metrics for all genres in the combination
        combined_metrics = pd.concat(genre_combination_metrics)
        
        # Calculate the mean metrics for this genre combination
        mean_metrics = combined_metrics.mean(axis=0)
        
        # Update the corresponding row in average_genre_metrics_df with the calculated mean metrics
        average_genre_metrics_df.loc[genre_combination] = mean_metrics

# Reset the index to make 'track_genre' a column
average_genre_metrics_df.reset_index(inplace=True)
average_genre_metrics_df.rename(columns={'index': 'track_genre'}, inplace=True)

# Print the DataFrame
print(average_genre_metrics_df)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

class SongRecommender:
    def __init__(self, sp, ultimate_df, song_df):
        self.sp = sp  # Spotify API client
        self.ultimate_df = ultimate_df
        self.song_df = song_df
        self.music_features_mbti = ['danceability_mean', 'energy_mean', 'loudness_mean', 'mode_mean',
                                    'speechiness_mean', 'acousticness_mean', 'liveness_mean', 'valence_mean',
                                    'tempo_mean']
        self.music_features_movie = ['danceability', 'energy', 'loudness', 'mode', 'speechiness', 'acousticness',
                                     'instrumentalness', 'liveness', 'valence', 'tempo']
        
        # Combine both sets of features
        self.music_features_combined = self.music_features_mbti + self.music_features_movie

        # Standard scaler for feature scaling
        self.scaler = StandardScaler()

        # Initialize NearestNeighbors with the correct number of features
        self.nearest_neighbors = NearestNeighbors(n_neighbors=5, algorithm='auto')


    def preprocess_content(self):
    # Extract MBTI-related music features
        features_mbti = self.ultimate_df[self.music_features_mbti].values
        
        # Extract movie-related music features
        features_movie = self.ultimate_df[self.music_features_movie].values
        
        # Concatenate the features
        combined_features = np.concatenate((features_mbti, features_movie), axis=1)
        
        # Convert to DataFrame to drop missing values
        combined_data = pd.DataFrame(combined_features, columns=self.music_features_mbti + self.music_features_movie)
        
        # Drop rows with missing values
        combined_data.dropna(inplace=True)
        
        # Extract features after dropping missing values
        cleaned_features = combined_data.values
        
        # Fit the StandardScaler on the cleaned features
        self.scaler.fit(cleaned_features)
        
        # Scale the features
        scaled_features = self.scaler.transform(cleaned_features)
        
        # Fit the NearestNeighbors model
        self.nearest_neighbors.fit(scaled_features)





    def recommend(self, character_name, num_recommendations=5):
        # Retrieve character data
        character_data = self.ultimate_df[self.ultimate_df['role'] == character_name]

        if character_data.empty:
            print(f"No data found for character: {character_name}")
            return []

        # Extract features for the character
        user_features = character_data[self.music_features_combined].values.reshape(1, -1)

        # Scale the features
        scaled_user_features = self.scaler.transform(user_features)

        # Find nearest neighbors
        _, top_indices = self.nearest_neighbors.kneighbors(scaled_user_features)

        # Retrieve track IDs of recommended songs
        track_ids = self.song_df.iloc[top_indices[0]]['track_id'].tolist()

        # Get track recommendations
        recommendations = self._get_track_recommendations(track_ids, num_recommendations)
        return recommendations

    def _get_track_recommendations(self, track_ids, num_recommendations):
        recommendations = []
        for track_id in track_ids:
            track_info = self.sp.track(track_id)
            track_name = track_info['name']
            artist_name = track_info['artists'][0]['name']
            recommendations.append(f"{track_name} by {artist_name}")
        return recommendations
```
